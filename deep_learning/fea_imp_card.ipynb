```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 设置绘图区域背景为白色
plt.rcParams['figure.facecolor'] = 'white'
import seaborn as sns

# 定义一个函数来绘制特征重要性的条形图
def plot_feature_importances(fea_imp, fn, ax, title):
    mean_abs_imp = np.abs(fea_imp).mean(axis=0)
    n = 30  # 选择前30个最重要的特征
    top_indices = np.argsort(-mean_abs_imp)[:n]
    feature_names = [fn[i] for i in top_indices]
    sns.barplot(x=mean_abs_imp[top_indices], y=feature_names, ax=ax)
    ax.set_title(title)
    ax.set_xlabel('Feature Importance (x1e-3)')
    ax.set_ylabel('Features')
    ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='lightgrey')
    # 将横坐标的值乘以1000，保留两位小数
    ax.set_xticklabels([f"{x*1000:.2f}" for x in ax.get_xticks()], rotation=45)
```

```python
fn_1 = pd.read_csv("/home/luojiawei/pengxiran_project/fn1.csv",header=0).iloc[:,2]
fn_2 = pd.read_csv("/home/luojiawei/pengxiran_project/fn2.csv",header=0).iloc[:,2]
fn_3 = pd.read_csv("/home/luojiawei/pengxiran_project/fn3.csv",header=0).iloc[:,2]
fn_4 = pd.read_csv("/home/luojiawei/pengxiran_project/fn4.csv",header=0).iloc[:,2]
fn_5 = pd.read_csv("/home/luojiawei/pengxiran_project/fn5.csv",header=0).iloc[:,2]
```

```python
fn_pre = fn_1.tolist() + fn_2.tolist()
fn_surg_end = fn_1.tolist() + fn_2.tolist()+ fn_3.tolist()
fn_post = fn_1.tolist() + fn_2.tolist()+ fn_3.tolist() + fn_4.tolist() + fn_5.tolist()
```

```python
y_name = "in_hospital_death"
fea_imp_pre = np.load(f'/home/luojiawei/pengxiran_project/结果文件夹/fea_imp_folder/fea_imp_{y_name}_pre.npy')
fea_imp_surg_end = np.load(f'/home/luojiawei/pengxiran_project/结果文件夹/fea_imp_folder/fea_imp_{y_name}_surg_end.npy')
fea_imp_post = np.load(f'/home/luojiawei/pengxiran_project/结果文件夹/fea_imp_folder/fea_imp_{y_name}_post.npy')

print("In-hospital death | absolute feature importace (Top 30)")
# 设置图形大小和布局
fig, axs = plt.subplots(1, 3, figsize=(20, 8))
# 绘制每个模型的特征重要性
plot_feature_importances(fea_imp_pre, fn_pre, axs[0], 'Just Before Surgery Feature Importance')
plot_feature_importances(fea_imp_surg_end, fn_surg_end, axs[1], 'Immediately After Surgery Feature Importance')
plot_feature_importances(fea_imp_post, fn_post, axs[2], '24 Hours After Surgery Feature Importance')

plt.tight_layout()
plt.show()
```
